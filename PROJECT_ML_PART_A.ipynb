### filling missing value 

```python
mean_length=dataset_no_dup["Length"].mean()
mean_diameter=dataset_no_dup["Diameter"].mean()
mean_height=dataset_no_dup["Height"].mean()
mean_whole_weight=dataset_no_dup["Whole weight"].mean()
mean_shucked_weight=dataset_no_dup["Shucked weight"].mean()
mean_viscera_weight=dataset_no_dup["Viscera weight"].mean()
mean_shell_weight=dataset_no_dup["Shell weight"].mean()
mean_rings=dataset_no_dup["Rings"].mean()
print(mean_length)
print(mean_diameter)
print(mean_height)
print(mean_whole_weight)
print(mean_shucked_weight)
print(mean_viscera_weight)
print(mean_shell_weight)
print(mean_rings)
```

## filing value

```python
dataset_no_dup.fillna({'Length':mean_length},inplace=True)
dataset_no_dup.fillna({'Diameter':mean_diameter},inplace=True)
dataset_no_dup.fillna({'Height':mean_height},inplace=True)
dataset_no_dup.fillna({'Whole weight':mean_whole_weight},inplace=True)
dataset_no_dup.fillna({'Shucked weight':mean_shucked_weight},inplace=True)
dataset_no_dup.fillna({'Viscera weight':mean_viscera_weight},inplace=True)
dataset_no_dup.fillna({'Shell weight':mean_shell_weight},inplace=True)
dataset_no_dup.fillna({'Rings':mean_rings},inplace=True)
```
